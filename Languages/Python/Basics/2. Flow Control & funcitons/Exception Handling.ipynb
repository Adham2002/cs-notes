## 3. Try and except


```python
print(21/0)
```

```python
## Try and except can be for error handling
## Try lets you test a block of code for errors
## Except block lets you handle the error should one be found
## Dividing with 0 as the denominator is impossible
## So the except block is executed in this instance

def impossible_maths():
    return 21 / 0
```

```python
## 
try:
    impossible_maths()
except:
    print('The maths was impossible')
```

```python
## This is possible maths

def possible_maths():
    return 0 / 21
```

```python
## Tries the impossible maths
## This raises an error 
## Instead excutes the possible maths

try:
    x = impossible_maths() ## will error
except:
    x = possible_maths() ## will run
    
print(x)
```

```python
## What if the exception is wrong too?

try:
    x = impossible_maths() 
except:
    x = impossible_maths()
```

> **Handled exception** - An error we account for and fix. Fixing can mean changing the inputs, limiting them, skipping etc.; anything we do to avoid running into the error

v.s


> **Unhandled exception** - An error we did not account for. Python will resort to one of its default exceptions to make us aware that the code is erronous and won't run

```python
## If an error is found we can use Exception to get details of the error
## Aliasing the Exception message as 'e'

try:
    impossible_maths()
except Exception as e:
    print(f'The maths was impossible: {e}')
```

You can specify the type of error you're expecting for more granularity:

```python
try:
    impossible_maths()
except ZeroDivisionError:
    print("I can't divide by 0")
except TypeError:
    print("I don't understand the specified type")
```

```python
## Let's change the order

try:
    impossible_maths()
except TypeError:
    print("I don't understand the specified type")
except ZeroDivisionError:
    print("I can't divide by 0")
```

<img src="https://www.section.io/engineering-education/exceptions-and-error/exception_hierarchy.png"/>

```python
## Python reads TOP to BOTTOM
## Will handle the first relevant error

try:
    impossible_maths()
except ArithmeticError:
    print("This error is Arithemtic! Check your calculations")
except ZeroDivisionError:
    print("You tried to divide by 0")
```

```python
## If we reverse the order, we get the ZeroDivisionError first

try:
    impossible_maths()
except ZeroDivisionError:
    print("You tried to divide by 0")
except ArithmeticError:
    print("This error is Arithemtic! Check your calculations")
```

You can add a generic except block to handle any type of error, but ONLY as the very final exception

```python
## Except is generic, it handles any type of exception
## Functionally the same as doing except BaseException

try:
    v = float('apple')
except ZeroDivisionError:
    print("You tried to divide by 0")
except ArithmeticError:
    print("This error is Arithemtic! Check your calculations")
except:
    print("This is a different type of error.")
```

```python
## What happens if we try to add an except branch in between?

try:
    v = float('apple')
except ZeroDivisionError:
    print("You tried to divide by 0")
except:
    print("This is a different type of error.")
except ArithmeticError:
    print("This error is Arithemtic! Check your calculations")
```

```python
## try: Block to try first
## except: Code executed if error is raised in try block
## else: Code executed if try block is successful
## finally: Last bit of code executed irregardless of any raised errors

try:
    impossible_maths()
    #possible_maths()
except:
    print('Maths was bad')
else:
    print('Maths was good')
finally:
    print('Yay we made it to the end')
```
