## Data Dictionary 

* Open: The price at the opening of the market
* High: The highest price point of that day
* Low: The lowest price point of that day
* Close: The closing price at end of day when market closes 
* Adj Close: Typically this is close pricing but taking into account the major movements in the market
* Volume: Total trades that day

## Creating a date/time range

```python
##

my_dates = ['2021-02-01', '2021-02-03', '2021-02-04', '2021-02-05', '2021-02-06']
steps_taken = [2309, 22039, 19000, 12900, 39000]
```

```python
##

steps_df = pd.DataFrame(zip(my_dates, steps_taken), columns=['Date', 'Steps'])
```

```python
##

steps_df.dtypes
```

```python
##
steps_df['Date'] = pd.to_datetime(steps_df['Date'])
steps_df.set_index('Date', inplace=True)
steps_df.sort_index(inplace=True)
```

```python
##

steps_df
```

```python
plt.figure(figsize=(12,9))
plt.plot(steps_df['Steps'])
```

```python
##
# https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#timeseries-offset-aliases

## YYYY-MM-DD
## ISO STANDARD FORMAT!


full_range = pd.date_range("2021-02-01", periods=6, freq="D")

full_range
```

```python

complete_df = pd.DataFrame(full_range, columns=['Date'])
complete_df.set_index('Date', inplace=True)
complete_df.head(6)
```

```python
complete_df = complete_df.join(steps_df, how='left')
complete_df
```

```python
##

complete_df['Steps'] = complete_df['Steps'].fillna(complete_df['Steps'].median())

##

plt.figure(figsize=(12,9))
plt.plot(complete_df['Steps'])
```

```python
complete_df.loc['2021-02-02']['Steps'] = 0

plt.figure(figsize=(12,9))
plt.plot(complete_df['Steps'])
```

```python
complete_df.loc['2021-02-02']['Steps'] = np.NaN

plt.figure(figsize=(12,9))
plt.plot(complete_df['Steps'])
```
