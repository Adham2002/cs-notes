```python
def advent1a(arr1 = None, arr2 = None, file_name = 'input.txt'):
    if not (arr1 or arr2):
        arr1 = []
        arr2 = []
        with open(file_name, "r") as f:
            for line in f:
                i, j = line.rstrip().split('   ')
                arr1.append(int(i))
                arr2.append(int(j))
    
    arr1_sorted = sorted(arr1)
    arr2_sorted = sorted(arr2)
    return sum([abs(arr1_sorted[i] - arr2_sorted[i]) for i in range(len(arr1))])
            


# Test 1 (given in q)
#advent1a([3,4,2,1,3,3], [4,3,5,3,9,3])
advent1a()
```

```python
def advent1b(file_name: str = 'input.txt'):
    d1: dict = {}
    d2: dict = {}
    with open(file_name, "r") as f:
        for line in f:
            i, j = line.rstrip().split('   ')
            d1[int(i)] = 1 if int(i) not in d1 else d1[int(i)] + 1
            d2[int(j)] = 1 if int(j) not in d2 else d2[int(j)] + 1
    
    nums = set.intersection(set(d1.keys()), set(d2.keys()))
    return sum(i * d1[i] * d2[i] for i in nums)

advent1b()
```
